```python
import pandas as pd
file_path = '/Users/xiaoyufan/Desktop/European_Ski_Resorts.csv'
df = pd.read_csv(file_path)
df
```

```python
countries = df['Country'].value_counts()
print(countries)
#countries.to_csv('/Users/xiaoyufan/Desktop/countries.csv', index=False)
```

```python
# List of countries
countries = ['Austria','France', 'Switzerland', 'Italy', 'Germany', 'Norway', 'Spain', 'Denmark', 'Sweden', 'Andorra', 'Slovakia']

# Filter and save data for each country
for country in countries:
    df_country = df[df['Country'] == country]
    file_name = f'/Users/xiaoyufan/Desktop/{country.lower()}.csv'  # Using lowercase country name for filename
    df_country.to_csv(file_name, index=False)
```
